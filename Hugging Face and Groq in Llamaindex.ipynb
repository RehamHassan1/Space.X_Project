##PDF RAG Example 2

```python
docs_cat = SimpleDirectoryReader(input_files=["/content/DISTRIBUTION TRANSFORMER.pdf"]).load_data()
```

```python
index_cat = VectorStoreIndex.from_documents(docs_cat)

retriever_cat = VectorIndexRetriever(index=index_cat)

query_engine_cat = RetrieverQueryEngine(retriever=retriever_cat)
```

```python
cat_prompt = f"Extract the parameters values stated next: {response}. and generate a response like this: Frequency: 50 Hz"
# test = "give a breif about the doc"

# Query the vectorstore index.
response_cat = query_engine_cat.query(cat_prompt)


print(response_cat)
```

```python
import pandas as pd
import re


response_text = response_cat.response  # Extract text from response object


# Extract key-value pairs using regex
data = []
for line in response_text.strip().split("\n"):
    match = re.match(r"(.+?):\s*(.+)", line)
    if match:
        key, value = match.groups()
        data.append({"Key": key.strip(), "Value": value.strip()})

# Convert to DataFrame
df = pd.DataFrame(data)

# Save to Excel
df.to_excel("extracted_data.xlsx", index=False)

print("Excel file 'extracted_data.xlsx' generated successfully!")
```
